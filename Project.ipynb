```python
import sqlite3 
import csv 
import pandas as pd
```

```python
con = sqlite3.connect('Prime_Tv_Shows_Data_set.db')
cursor = con.cursor()
```

```python
create_table = """CREATE TABLE Prime_Shows(
    S_no INTEGER PRIMARY KEY,
    Name_of_the_Show TEXT NOT NULL,
    Year_of_release TEXT NOT NULL,
    No_of_Seasons INTEGER NOT NULL,
    Language TEXT NOT NULL,
    Genre TEXT NOT NULL,
    IMBD_rating FLOAT NOT NULL,
    Age_of_Viewers INTEGER NOT NULL
);
"""

cursor.execute(create_table)
```

```python
df = pd.read_csv(r'C:\Users\dhyey\Documents\Python111\Project 6\Prime_TV_Shows_Data_set.csv', encoding='latin1')

# Insert DataFrame into SQLite table
df.to_sql('hotel_data',con, if_exists='append', index=False)
```

```python
# Fetch table schema
cursor.execute("PRAGMA table_info(Prime_Shows)")
schema = cursor.fetchall()

# Print table schema
for column in schema:
    print(column)
```
